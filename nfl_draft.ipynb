```python
data = df[df['pos'] == 'QB']

sns.boxplot(
    data=data,
    x="rnd", y="pass_yds"
)
```

```python
ax = sns.barplot(data=df, x='year', y='rush_yds')

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=8)
plt.tight_layout()
plt.show()
```
