```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(".data_sets/nfl_draft.csv")
df = df[(df['tm'] == "STL") | (df['tm'] == "RAM")]
df['prem'] = df['pos'].isin(['QB', 'DE', "T", "WR", "DB"])
```

```python
palette_color = sns.color_palette()

top = df[df['rnd'] == 1]
data = top.groupby("pos").size().to_frame("count").reset_index()
  
plt.pie(data["count"], labels=data["pos"], colors=palette_color, autopct='%.0f%%')
plt.title("Postions taken Round 1 by the Rams (1985-2015)")
  
plt.show()
```

```python
fig, axes = plt.subplots(1,2)
fig = plt.figure(figsize=(10, 10))

sns.barplot(
    data=df,
    x="rnd",
    y="ap1",
    ax=axes[0],
    estimator=sum
)

sns.barplot(
    data=df,
    x="rnd",
    y="pb",
    ax=axes[1],
    estimator=sum
)
axes[0].set_title("Rams Total All Pros by round")
axes[0].set_xlabel("Round")
axes[0].set_ylabel("All Pros")

axes[1].set_title("Rams Total Pro Bowls by round")
axes[1].set_xlabel("Round")
axes[1].set_ylabel("Pro Bowls")
```

```python
data = df[df['prem']]

sns.lineplot(data=data, x='rnd', y='carav', hue='pos', errorbar=None)
plt.xlabel("Round")
plt.ylabel("Career AV")
plt.title("Rams Career AV by round for premium positions")
plt.show()
```

```python
sns.scatterplot(data=df, x='pick', y='carav', hue='pos')
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.xlabel("Pick")
plt.ylabel("Career AV")
```
